```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn import linear_model
import openpyxl
import warnings
```

```python
d= pd.read_csv("D:\\python\\datasets\\canada_per_capita_income.csv")
df = pd.DataFrame(d)
df.head()
df.describe()
```

```python
%matplotlib inline


# parcapita = df('per capita income (US$)')

xaxis = plt.xlabel("year",fontsize = "20")
yaxis = plt.ylabel("percapita",fontsize = "20")

plt.scatter(df.year,df.percapita,marker="+")




```

```python
reg= linear_model.LinearRegression()
# reg.fit(df[['Area']], df.Price)
reg.fit(df[['year']], df.percapita)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
xaxis = plt.xlabel("year",fontsize = "20")
yaxis = plt.ylabel("percapita",fontsize = "20")

plt.scatter(df.year,df.percapita,marker="+",color='red')

plt.plot(df.year,reg.predict(df[['year']]),color='blue',)
```
