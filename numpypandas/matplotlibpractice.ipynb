MARKERS

marker is the fuction to mark th points on line

mfc is full color
mec is the edge color

```python
x=np.arange(1,10,2)
y=[i**2 for i in x]
p = plt.plot(x,y,marker="*",ms="15",mfc="y",mec="k")
plt.title("Simple Line Plot")
plt.xlabel("x-axis")
plt.xlabel("y-axis")
plt.show()
```

line on graph

types of line style

solid, dotted, dashed, dsshdot, name

line style - ls use cheyali
line width - lw use cheyali
line color - c use cheyali


```python
x=np.arange(1,10,2)
y1=[i**2 for i in x]
y2=x**3
p = plt.plot(x,y1,y2,marker="*",ms="15",mfc="y",mec="k",ls="dotted", lw= "5", c="y")
plt.title("Simple Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

Font properties

```python
#x and y axis and cordinated
x=np.arange(1,10,2)
y1=[i**2 for i in x]
y2=x**3
#fonts
f1={"family":"Georgia","size":20,"color":'k'}
f2={"family":"calibri","size":20,"color":'y'}
f3={"family":"arial","size":20,"color":'c'}

p = plt.plot(x,y1,y2,marker="*",ms="15",mfc="y",mec="k",ls="dotted", lw= "5", c="y")
plt.title("Simple Line Plot",fontdict=f1)
plt.xlabel("x-axis",fontdict=f2)
plt.ylabel("y-axis",fontdict=f3)
plt.show()
```

grid lines in graph

```python
#x and y axis and cordinated
x=np.arange(1,20,2)
y1=[i+140 for i in x]
y2=x+3
#fonts
f1={"family":"Georgia","size":20,"color":'k'}
f2={"family":"calibri","size":20,"color":'y'}
f3={"family":"arial","size":20,"color":'c'}

#grid lines
plt.grid(axis="both",ls="solid", lw= "2", c="k")

p = plt.plot(x,y1,y2,marker="*",ms="15",mfc="y",mec="k",ls="dotted", lw= "5", c="y")
plt.title("Simple Line Plot",fontdict=f1)
plt.xlabel("x-axis",fontdict=f2)
plt.ylabel("y-axis",fontdict=f3)
plt.show()
```

sub plots

syntax

subplot(R,C,I) #R=row, c=column , I = index


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x=[0,1,2,3,4]
y1=[50,20,30,10,40]
y2=[40,50,20,30,10]

#fonts
f1={"family":"Georgia","size":20,"color":'k'}
f2={"family":"calibri","size":12,"color":'y'}
f3={"family":"arial","size":12,"color":'c'}
f4={"family":"Georgia","size":20,"color":'r'}

#grid lines
#plt.grid(axis="both",ls="solid", lw= "2", c="k")

plt.suptitle("Main plots",fontdict=f1)
#1nd sub plot
plt.subplot(2,2,1)

p = plt.plot(x,y1,marker="*",ms="15",mfc="y",mec="k",ls="solid", lw= "2")
plt.title("Sub plot 2",fontdict=f1)
plt.xlabel("x-axis",fontdict=f2)
plt.ylabel("y-axis",fontdict=f3)

#2nd sub plot
plt.subplot(2,2,2)

p = plt.plot(x,y2,marker="*",ms="15",mfc="y",mec="k",ls="solid", lw= "2")
plt.title("Sub plot 2",fontdict=f1)
plt.xlabel("x-axis",fontdict=f2)
plt.ylabel("y-axis",fontdict=f3)


#3nd sub plot
plt.subplot(2,2,3)

p = plt.plot(x,y2,marker="*",ms="15",mfc="y",mec="k",ls="solid", lw= "2")
plt.title("Sub plot 3",fontdict=f1)
plt.xlabel("x-axis",fontdict=f2)
plt.ylabel("y-axis",fontdict=f3)


#4nd sub plot
plt.subplot(2,2,4)

p = plt.plot(x,y2,marker="*",ms="15",mfc="y",mec="k",ls="solid", lw= "2")
plt.title("Sub plot 4",fontdict=f1)
plt.xlabel("x-axis",fontdict=f2)
plt.ylabel("y-axis",fontdict=f3)
```

Legend

```python
x=np.arange(0,5,1)
y1=x**2
y2=x**3

#plt.plot(x,y1,y2)
#plt.legend(['Squares','Cubes'])

plt.plot(x,y1,label="Squares")
plt.plot(x,y2,label="Cubes")
plt.legend(loc="upper left",framealpha=1,facecolor="yellow",edgecolor="black",fancybox=True,shadow=True,fontsize=12)

plt.show()

```

scatter plots

```python
import random


x=np.arange(1,11,1)
y=np.arange(10,110,10)
random.shuffle(y)
print(x,y)
plt.scatter(x,y,color="red",marker="s",s=200,linewidths=1,alpha=0.5)

x1=np.arange(1,11,1)
y1=np.arange(10,110,10)
random.shuffle(y1)
print(x1,y1)
plt.scatter(x1,y1,color="Green",marker="^",s=200,linewidths=1,alpha=0.9)

plt.show()

```

color bar

```python
x=np.arange(1,6,1)
y=np.arange(10,60,10)
random.shuffle(y)
colors=[70,20,10,50,100]
sizes=[200,230,300,250,150]
print(x,y)
plt.scatter(x,y,c=colors,marker="^",s=sizes,linewidths=1,alpha=0.5,cmap="rainbow")

plt.colorbar()

plt.show()
```
