```python
import matplotlib as mat
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings as war
import seaborn as sns
print("matplotlib vers is ",mat.__version__)
print("numpy vers is ",np.__version__)
print("Pandas vers is ",pd.__version__)
print("seaborn vers is ",sns.__version__)
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips.head()
sns.barplot(x='day',y='total_bill',hue='sex',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',estimator=np.median,data=tips)
```

```python
sns.barplot(x='day',y='total_bill',ci=60,capsize=0.3,saturation=0.7, data=tips)
```

```python
sns_plot = sns.barplot(x='day',y='total_bill',data=tips)
```
